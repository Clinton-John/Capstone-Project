```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('/content/jumia_articles.csv')
```

```python
data.head()
```

```python
## Finding Number of rows and columns in the dataset
rows_count, columns_count = data.shape
print(f"There are {rows_count} rows, and {columns_count} columns")
```

Data Preprocessing

```python
data.info()
```

```python
##from the above, all rows don't have the number of ratings. This should be dropped.Creating a new dataframe (data1)
data1 = data.drop('Number of Ratings', axis=1)
data1.head()
```

```python
#creatin a new column known as brand and adding the first word in every Article Name to it.

data1['Article Brand'] = data1['Article Name'].str.split().str[0]

#Placing the Article Brand column after the first column
cols = list(data1.columns)
cols.insert(1, cols.pop(cols.index('Article Brand')))  # Move 'brand' to the second position

# # Reorder the DataFrame columns
data1 = data1[cols]
data1.head()
```

Creating the Article Rating Column to only have the Ratings number and 0 if there is none

```python
data1['Article Rating'] = data1['Article Rating'].str.split().str[0]
data1['Article Rating'] = data1['Article Rating'].fillna('0')
```

```python
data1.head()
```

```python
data1.info()
```

```python
data1.head()
```

```python
#Discount Percentage Column to be 0 %
data1['Discount Percentage'] = data['Discount Percentage'].fillna('0%')
data1.head()
```

```python
## Filling the Initial Price to be the same as the current price if it is null
data1['Initial Price'].fillna(data1['Current Price'], inplace=True)
data1.head()
```

```python
## Removing all percentage signs on the Discount percentage column to make it an integer
data1['Discount Percentage'] = data1['Discount Percentage'].str.rstrip('%')
data1.head()
```

```python
## removing white spaces from the Current Price column and the Initial Price Column
data1['Current Price'] = data1['Current Price'].str.strip()
data1['Initial Price'] = data1['Initial Price'].str.strip()
data1.head()
```

```python
# Removing the Sh at the begining of each and every row of the Current Price and Initial Program
data1['Current Price'] = data1['Current Price'].str.lstrip('Sh')
data1['Initial Price'] = data1['Initial Price'].str.lstrip('Sh')
data1.head()
```

```python
# Removing all the commas in the initial Price and Current price columns
data1['Current Price'] = data1['Current Price'].str.replace(',', '')
data1['Initial Price'] = data1['Initial Price'].str.replace(',', '')
data1.head()
```

```python
#Current Price has some values which are in ranges and need to be removed. Check their count
# data1['Current Price'].count()
range_df = data1[data1['Current Price'].str.contains('-')]
range_count = data1['Current Price'].str.contains('-').sum()
print(f"There are {range_count} prices with a range of prices")
range_df.head()
range_df.shape
# range_df['location'].value_counts()
```

```python
data1['Current Price'] = data1['Current Price'].astype(str)
data1['Initial Price'] = data1['Initial Price'].astype(str)
```

```python
##Dropping the rows that have their initial price and current price as a range
data1.drop(data1[data1['Current Price'].str.contains('-')].index, inplace=True)
data1.drop(data1[data1['Initial Price'].str.contains('-')].index, inplace=True)

data1.head()
```

```python
data1.shape
```

Coverting all columns to respective datatypes
Article Name - String

Article Brand - String

Current Price - Integer

Iniaital Price - Integer

Discount Percentage - Integer

Article Rating - Float

location - string

category - string

```python
data1['Article Name'] = data1['Article Name'].astype(str)
data1['Article Brand'] = data1['Article Brand'].astype(str)
data1['Current Price'] = data1['Current Price'].astype(int)
data1['Initial Price'] = data1['Initial Price'].astype(int)
data1['Discount Percentage'] = data1['Discount Percentage'].astype(int)
data1['Article Rating'] = data1['Article Rating'].astype(float)
data1['location'] = data1['location'].astype(str)
data1['category'] = data1['category'].astype(str)
```

```python
#Convert the data1 into a csv file ready for transfer into the local database
data1.to_csv('jumia_articles_prep.csv', index=False)
```
